# Data Visualization

Univariate Analysis ([Plotting Tutorial](https://www.kaggle.com/residentmario/univariate-plotting-with-pandas))
*    Plot the distribution of numerical variables with [histograms](https://seaborn.pydata.org/tutorial/distributions.html#plotting-univariate-histograms) or [density plots](https://seaborn.pydata.org/tutorial/distributions.html#kernel-density-estimation).
*    Look for outliers of numerical variables using [violin plots](https://seaborn.pydata.org/generated/seaborn.violinplot.html) or [box plots](https://seaborn.pydata.org/generated/seaborn.boxplot.html).
*    Understand the numerical distribution of categorical variables by calling the `value_counts` **method** on a single Series column.
*    Plot the distribution of categorical variables with [bar charts](https://seaborn.pydata.org/generated/seaborn.barplot.html#seaborn.barplot) (no pie charts please, we hate pie charts).


```python
fig, ax = plt.subplots(figsize = (15, 5))

sns.set_palette("crest")

sns.barplot(x = 'index',
            y = 'genre',
            data = pd.DataFrame(spotify_df['genre'].value_counts()).reset_index())

plt.xlabel("genre")
plt.ylabel("count")
plt.title("Count of Songs in Each Genre in the Top Songs of 2010-2019")
plt.show()
```

Interestingly, but also expectedly, pop is by far the most popular genre of song in the top 100 spotify songs of each year. Hip hop and rap are next but even when considered together are still not as prevalent as pop. And almost half the genre bars look neglible in this graph.

```python
fig, ax = plt.subplots(3, 2, figsize = (12, 8))

pal = sns.color_palette('crest', 8)

# create histograms
sns.histplot(spotify_df['pop'], ax = ax[0][0], color = pal[0])
sns.histplot(spotify_df['bpm'], ax = ax[0][1], color = pal[1])
sns.histplot(spotify_df['val'], ax = ax[1][0], color = pal[2])
sns.histplot(spotify_df['dur'], ax = ax[1][1], color = pal[3])
sns.histplot(spotify_df['dnce'], ax = ax[2][0], color = pal[4])
sns.histplot(spotify_df['dB'], ax = ax[2][1], color = pal[5])

plt.tight_layout()
```

Here we plot the distribution of many of the most important numerical variables. Most seem to be normally distributed, even if they do have some skewed values. There evidently seem to be some really long songs and some really loud songs.

Bivariate Analysis ([Plot Ideas](https://www.kaggle.com/residentmario/bivariate-plotting-with-pandas))

    * [Scatter Plots](https://seaborn.pydata.org/tutorial/relational.html#relating-variables-with-scatter-plots)
    * [Line Plots](https://seaborn.pydata.org/tutorial/relational.html#emphasizing-continuity-with-line-plots)
    * [Regression Plots](https://seaborn.pydata.org/tutorial/regression.html)

```python
fig, ax = plt.subplots( figsize = (12, 5))

sns.set_palette("Reds", 12)

# Create the data for the std. deviation plot
pop_std = spotify_df.groupby('genre').agg('std').sort_values('pop')['pop']
pop_std = pd.DataFrame(pop_std).reset_index()

# Create the data for mean plot.
pop_mean = spotify_df.groupby('genre').agg('mean').sort_values('pop', ascending = False)['pop']
pop_mean = pd.DataFrame(pop_mean).reset_index()

# Plot means
sns.barplot(x = 'genre',
            y = 'pop',
            data = spotify_df,
            dodge = False,
            order = spotify_df.groupby('genre').agg('mean').sort_values('pop').index,
            ax = ax
            )

plt.ylabel("Popularity")
plt.ylim(57,83)
plt.title("Popularity of Each Genre")
plt.show()
```

Popularity in this dataset is an index of how well the song performed commercially, not a ranking. The height of this graph shows the average popularity across each genre, and the bar on top shows the variances. So we can see for example pop music tends to be popular but not the most popular, and tends to have a small variance.

A few things I noticed:
- Dance has the lowest average popularity, but also has a really high variance. This seems to suggest that many dance songs that made it on the list are not so popular, but popular dance songs have potential to become really popular.
- Almost oppositely latin songs that landed on the year end list tend to be very popular with a small variance, so latin songs that get popular tend to get *really* popular.
- Despite (or maybe in spite) of being the overwhelming most represented genre in the dataset, pop music tends also to have the tightest variance. My guess is that maybe there are so many pop songs they inevitably tend to clump close to an average popularity, which fits in with the fact that pop is about dead center in terms of popularity. 

```python
fig, ax = plt.subplots(figsize = (12, 5))

# pal = sns.color_palette("Blues", as_cmap=True)
sns.set_palette("ch:start=.4,rot=-.3", 12)

# Create the data for mean plot.
dB_std = spotify_df.groupby('genre').agg('mean')['dB']
dB_std = pd.DataFrame(dB_std).reset_index()
dB_std['opposites'] = -1 * dB_std['dB']

o = ['dance', 'rap', 'rock', 'edm/house', 'hip hop', 'pop soul', 'indie', 'country', 'pop', 'latin', 'r&b', 'other']

# Plot means
barplot = sns.barplot(y = 'dB',
                      x = 'genre',
                      data = dB_std,
                      dodge = False,
                      order = o,
                      )

barplot = sns.barplot(y = 'opposites',
                      x = 'genre',
                      data = dB_std,
                      order = o,
                      dodge = False)

plt.ylabel("Loudness (Decibels)")
plt.title("Average Loudness of Each Genre")
plt.show()
```

For this plot I wanted to make it look like a little audio graph and appropriately it shows the relative loudness of each genre.

```python
# create the group by object dataset.
by_year = spotify_df.drop(['title', 'artist', 'top genre', 'added', 'artist type', 'genre'], 
                          axis = 1).groupby('top year').agg(['mean'])

fig, ax = plt.subplots(2, 2, figsize = (12, 8))

# Set a style and color palette.
plt.style.use('fivethirtyeight')

pal = sns.color_palette('mako', 4)

# Plot each metric
sns.lineplot(x = list(by_year.index),
             y = list(by_year['dur'].reset_index()['mean']),
             color = pal[0],
             ax = ax[0][0])

sns.lineplot(x = list(by_year.index),
             y = list(by_year['bpm'].reset_index()['mean']),
             color = pal[1],
             ax = ax[0][1])

sns.lineplot(x = list(by_year.index),
             y = list(by_year['nrgy'].reset_index()['mean']),
             color = pal[2],
             ax = ax[1][0])

sns.lineplot(x = list(by_year.index),
             y = list(by_year['dnce'].reset_index()['mean']),
             color = pal[3],
             ax = ax[1][1])

# Add plot labels
ax[0][0].set(ylabel = "Duration")
ax[0][1].set(ylabel = "Speed (bpm)")
ax[1][0].set(ylabel = "Energy")
ax[1][1].set(ylabel = "Danceability")
fig.suptitle("Various Metrics of Top Spotify Songs Over Time")

plt.show()
```

Here we can see a little more clealy the dynamics of how these metrics have changed over time. While speed and energy have gone down, danceability has gone up. Songs also seem to have gotten shorter over the decade.

```python
sns.set_palette("Greens", 2)
sns.pairplot(spotify_df[['bpm', 'nrgy', 'dB', 'val', 'acous', 'pop']])
plt.show()
```

For my last plot I created a scatterplot matrix for a few of the variables that seemed importat. `bpm` is the speed of the song, `nrgy` is a measure of how energetic the song is, `dB` is the loudness of each song in decibels, `val` is a measure of how positive the song is, `acous` is a measure of how acoustic a song is, and `pop` is a measure of the popularity of each song.

Some of the relationships I noticed are:
- Popularity and acoustic seem to be slightly positively correlated.
    This seems to suggest that less popular songs are less likely to be acousitc? I.e. more acoustic songs have a slightly higher potential to become popular?
- Popularity and loudness seem to be negatively correlated, as do Popularity and Energy.
    This indicates that quieter and less energetic songs seem to be less likely to become popular.
- Energy and val seem to have a slight positive correlation as well, so songs with more energy seem to also be more positive. This kind of makes sense with what we would expect, as a less positive song would tend to be be slower.
- Loudness and energy are positively correlated. Higher energy songs tend to be louder.
